```python
import pandas as pd
import numpy as np
import plotly.figure_factory as ff
import scipy.cluster.hierarchy as sch
```

```python
df=pd.read_csv("E:\\ML Training\\Unsupervised_Learning\\Wholesale customers data.csv")
```

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
scaled_data = StandardScaler().fit_transform(df)
```

```python
df.rename(columns={'Annual Income (k$)':'Income'
,'Spending Score (1-100)':'Score'},inplace=True)
```

```python
df
```

```python
plt.figure(figsize=(10,15))
sch.dendrogram(sch.linkage(scaled_data,"ward"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=2)
df['group']=model.fit_predict(scaled_data)
```
