```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import plotly.express as px
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

### Load Data Set




```python
df=pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/oil_dataset.csv')
```

```python
df
```

```python
df.plot(x='Mendacium',y='Depth',kind='kde')
```

```python
px.scatter_3d(df,x='Mendacium',y='Depth',z='Price')
```

```python
x=df[['Mendacium','Depth']]
y=df['Price']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
model=DecisionTreeRegressor(max_depth=6)
model.fit(xtrain,ytrain)
```

```python
print(model.score(xtest,ytest)*100)
```

```python
pred=model.predict(x)
```

```python
print("Mean Absolute Eroor:",mean_absolute_error(y,pred))
print("Mean_squared_error:",mean_squared_error(y,pred))
```
